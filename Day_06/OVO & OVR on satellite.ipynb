```python
# Import necessary libraries
from sklearn.linear_model import *
from sklearn.metrics import *
from sklearn.linear_model import *
from sklearn.model_selection import *
from sklearn.compose import *
from sklearn.svm import SVC
from sklearn.model_selection import *
from sklearn.model_selection import *
from sklearn.discriminant_analysis import *
from sklearn.preprocessing import *
from sklearn.neighbors import *
from sklearn.naive_bayes import *
from sklearn.pipeline import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
from sklearn.linear_model import *
import os

warnings.simplefilter('ignore')

os.chdir('D:/Datasets')

satellite = pd.read_csv('Satellite.csv',sep=';')
satellite
```

```python
# Import the LabelEncoder from sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Encode the 'classes' column in the 'satellite' DataFrame
satellite['classes'] = le.fit_transform(satellite['classes'])

# Print the classes (unique labels) that were encoded
print(le.classes_)
```

```python
x = satellite.drop('classes',axis=1)
y = satellite['classes']
```

```python
# Split the dataset into training and testing sets
# x and y are the feature and target variables, respectively
# test_size=0.3 means 30% of the data will be used for testing
# stratify=y ensures that the class distribution in the training and testing sets is similar to the original dataset
# random_state=24 sets the random seed for reproducibility
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=24)

# Create a Support Vector Machine (SVM) classifier
# probability=True enables probability estimates for the predictions
# random_state=24 sets the random seed for reproducibility
# decision_function_shape='ovo' sets the decision function shape to one-vs-one
svm = SVC(probability=True, random_state=24, decision_function_shape='ovo')

# Train the SVM classifier on the training data
svm.fit(x_train, y_train)

# Use the trained SVM classifier to make predictions on the test data
y_pred = svm.predict(x_test)

# Calculate the accuracy score
# y_test is the true target values
# y_pred is the predicted target values
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy score
print(accuracy)
```

```python
# Use the trained SVM classifier to predict the probabilities for the test data
y_pred_prob = svm.predict_proba(x_test)

# Calculate the log loss (cross-entropy loss)
# y_test is the true target values
# y_pred_prob is the predicted probabilities
log_loss_score = log_loss(y_test, y_pred_prob)

# Print the log loss score
print(log_loss_score)
```

# OVR & OVR (GridSearch)

```python
params = {
    'kernel': ['rbf', 'linear'],
    'C': np.linspace(0.01, 4, 10),
    'gamma': np.linspace(0.01, 4, 10),
    'decision_function_shape': ['ovo', 'ovr']
}

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)
svm = SVC(probability=True, random_state=24)
gcv = GridSearchCV(svm, param_grid=params, verbose=3, cv=kfold, scoring='accuracy')
gcv.fit(x, y)

print(gcv.best_params_)
print(gcv.best_score_)
```

# Randomize search

```python
params = {
    'kernel': ['rbf', 'linear'],
    'C': np.linspace(0.01, 4, 10),
    'gamma': np.linspace(0.01, 4, 10),
    'decision_function_shape': ['ovo', 'ovr']
}

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)
svm = SVC(probability=True, random_state=24)
rgcv = RandomizedSearchCV(svm, param_distributions=params, verbose=3,random_state=24, cv=kfold, scoring='accuracy')
rgcv.fit(x, y)

print(rgcv.best_params_)
print(rgcv.best_score_)
```
