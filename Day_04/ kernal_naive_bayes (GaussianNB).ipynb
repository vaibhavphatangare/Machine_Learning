```python
# Import necessary libraries
from sklearn.linear_model import *
from sklearn.metrics import *
from sklearn.linear_model import *
from sklearn.model_selection import *
from sklearn.model_selection import *
from sklearn.model_selection import *
from sklearn.preprocessing import *
from sklearn.neighbors import *
from sklearn.naive_bayes import *
from sklearn.pipeline import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
from sklearn.linear_model import *
import os
```

```python
# Ignore warnings
warnings.simplefilter("ignore")

# Load the training data into a DataFrame named "mohs"
hr = pd.read_csv("HR_comma_sep.csv", index_col=0)

dum_hr = pd.get_dummies(hr,drop_first = True)
x = dum_hr.drop('left', axis=1)
y = dum_hr['left']
hr
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=24)

# applying GaussianNB()
nb = GaussianNB()
nb.fit(x_train,y_train)
y_pred = nb.predict(x_test)
y_pred_proba = nb.predict_proba(x_test)
print("Confusion matrix: \n",confusion_matrix(y_test,y_pred))
print('r2 score(without scaling): ',r2_score(y_test,y_pred))
```

# kfold


```python
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=24)
nb= GaussianNB()
#with accuracy
results = cross_val_score(nb,x,y,cv=kfold)
print('accuracy: ',results.mean())
#with log loss
results = cross_val_score(nb,x,y,cv=kfold,scoring='neg_log_loss')
print('log loss: ',results.mean())
```
